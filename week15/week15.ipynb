# 当前定位获取

```python

import requests

# 高德开发者Key,需要到高德开发者后台获取
gaode_key = 'e12b6979c90c84b45aa6a520367160b7'

def get_location():
    url = f'https://restapi.amap.com/v3/ip?key={gaode_key}'
    res = requests.get(url)
    if res.status_code == 200:
        print(res.json())
        data = res.json()
        return "经度是：" + data['rectangle'].split(',')[1] + ",维度是：" + data['rectangle'].split(',')[0]

if __name__ == '__main__':
    get_location()
```

# 目的地定位获取

```python
import requests

gaode_key = 'e12b6979c90c84b45aa6a520367160b7'

address = input('请输入您要查询的地址或地理信息:')
url = f'https://restapi.amap.com/v3/geocode/geo?key={gaode_key}'
payload = {
    'address': address,
    'output': 'JSON'
}
res = requests.get(url, params=payload)
aim_data = res.json()

aim_lon = aim_data['geocodes'][0]['location'].split(',')[0] #目的地经度
aim_lat = aim_data['geocodes'][0]['location'].split(',')[1] #目的地纬度
print(aim_lon,aim_lat)
aim_location = str(aim_data['geocodes'][0]['location'])
aim_location
```
